```python
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
import pylab
```

```python
bb_df = pd.read_csv('baseball_data.csv')
```

```python
# 数据完整性
# 一共1157个观测值，没有缺失数据
bb_df.info()
```

```python
bb_df[:10]
```

```python
# 用手习惯分布
# 64% 用右手， 27%左手， 9% 双手
handedness = bb_df.groupby('handedness')
handedness.size().plot(kind='pie', autopct = '%.0f%%')
plt.ylabel('count')
plt.title('Handness Vs Count')
plt.show()
```

```python
# 身高分布
height = bb_df.groupby('height')
height.size().sort_values().plot(kind='barh')
plt.title('height vs count')
plt.show()
```

```python
# 体重分布
weight = bb_df.groupby('weight')
weight.size().plot(kind='line')
plt.xlim(140,250,50)
plt.title('weight vs count')
plt.ylabel("count")
plt.show()
```

```python
%matplotlib inline

plt.xlabel('avg')
plt.ylabel('HR')
plt.title('avg VS HR')
plt.scatter(bb_df['avg'], bb_df['HR'])
```

```python
# 身高分组
bins_height = np.arange(60, 85, 1)
bb_df['height_group'] = pd.cut(bb_df['height'], bins_height)
bb_df['height'].describe()
```

```python
# 体重分组
bins_weight = np.arange(135,250, 5)
bb_df['weight_group'] = pd.cut(bb_df['weight'], bins_weight)
```

```python
# avg 分组
bins_avg = np.arange(0, 0.4, 0.02)
bb_df['avg_group'] = pd.cut(bb_df['avg'], bins_avg, right=False)
bb_df['avg'].describe()
```

```python
bb_df.groupby(['height_group'])['HR'].sum().plot(kind = 'bar')
```

```python
bb_df.groupby(['weight_group'])['HR'].sum().plot(kind = 'bar')
```

```python
bb_df.groupby('avg_group')['HR'].sum().plot(kind = 'bar')
```

```python
bb_df.to_csv('Baseball_new.csv')
```
